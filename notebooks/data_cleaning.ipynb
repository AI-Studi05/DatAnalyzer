## Data train/test split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler, PolynomialFeatures
```

```python
X=data_wine_red.iloc[:,:-1]
y=data_wine_red["quality"]
```

```python
y
```

```python
def split_data(X,y,random_state=None):
    X_train, X_test, y_train, y_test= train_test_split(X,y,random_state=random_state,test_size=0.5)
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test=split_data(X,y,random_state=42)
```

## Preprocessing 

### Min-max scaler

```python
def minmaxscaler():
    min_max_scaler = MinMaxScaler()
    X_train_minmax=min_max_scaler.fit(X_train)
    X_train_minmax=X_train_minmax.transform(X_train_minmax)
    return X_train_minmax
```

### z-normalisation with standard scaler

```python
def z_normalisation():
    scaler = StandardScaler()
    X_train_minmax=scaler.fit(X_train)
    X_train_minmax=X_train_minmax.transform(X_train_minmax)
    return X_train_minmax
```

```python
def polynomialscaler():
    polynomial_scaler=PolynomialFeatures()
    X_train_minmax=polynomial_scaler.fit(X_train)
    X_train_minmax=X_train_minmax.transform(X_train_minmax)
    return X_train_minmax
```
